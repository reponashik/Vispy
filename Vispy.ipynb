```python
import numpy as np 
import vispy
from vispy import app, visuals, scene
import matplotlib.pylab as plt 
```

```python
# required :
# https://vispy.org/installation.html
```

```python
n = 60 
x = np.sin(np.linspace(-2,2,n)* np.pi)
y = np.cos(np.linspace(-2,2,n)* np.pi)
z = np.linspace(-2,2,n)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x,y,z)
plt.show()
```

```python
plot3D = scene.visuals.create_visual_node(visuals.LinePlotVisual)

canvas = scene.SceneCanvas()

view = canvas.central_widget.add_view()
view.camera = 'turntable'
view.camera.fov = 45
view.camera.distance = 6

pos = np.c_[x,y,z] #numpy concatenation

plot3D(pos, width=30.0,color='red',symbol='o',edge_color='w',face_color=(0.2,0.2,1,0.8), parent=view.scene)
canvas.show()
app.run()
```
